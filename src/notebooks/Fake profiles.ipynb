## Creating data with The Synthetic Data Vault (SDV)

### Loading Gaussian Copula model for single table data

```python
from sdv.tabular import GaussianCopula
model = GaussianCopula()
model.fit(df_real_deal)
```

### Selecting the numbers of observations 

```python
sample = model.sample(30000)
# sample.selection_map.value_counts()
sample
```

### Evaluating how good is the data created

```python
from sdv.evaluation import evaluate
evaluate(sample, df_real_deal, metrics=['CSTest', 'KSTest'], aggregate=False)
```

```python
overall_score = evaluate(sample, df_real_deal)
```
