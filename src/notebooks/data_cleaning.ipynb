```python
import pandas as pd

pd.set_option('display.max_colwidth', None ) #50 is the default
```

```python
df = pd.read_csv('data/puntos_interes.csv', index_col= 0)

df_rutas = pd.read_csv('data/routes.csv', index_col= 0)
```

```python
df_rutas = df_rutas.rename({'description':'resume'}, axis=1)

df_rutas = df_rutas.reset_index().rename({'index':'route_id'}, axis=1)
```

```python
df_rutas['columna1'] = df_rutas.types.str.replace("[",'').str.replace(']','').str.split(',')

for i in range(len(df_rutas.columna1)):
    df_rutas['transport'] = df_rutas.columna1[i][0]

for i in range(len(df_rutas.columna1)):
    df_rutas['type'] = df_rutas.columna1[i][1]

```

```python
df_rutas = df_rutas.drop(['types','columna1'],axis=1)
```

```python
df_rutas['duration'] = [120, 30, 60, 60, 60, 360, 105, 105, 105, 240, 180, 90, 90, 120, 60, 60, None, 420, 360, None, 180, 420, 240, None] 
```

```python
names = list(df.groupby('nombre_ruta').count().index)

names.append('Ruta València en bicicleta')

indices = [0, 2, 1, 3, 4, 6, 7, 8, 9, 10, 11, 5, 12, 23, 13, 14, 15, 16, 19, 18, 20, 21, 22,17]

```

```python
df_2 = pd.DataFrame({'names':names, 'indices':indices})

df_2 = df_2.sort_values(by='indices')

names_2 = list(df_2.names)
```

```python
df_rutas['name'] = names_2

df_rutas
```

```python
descripcion = list(df.groupby('descripcion_ruta').count().index)

descripcion.append("Si apostes pel transport sostenible i t'agrada anar amb bicicleta, has de saber que València aplega condicions ideals per al turisme ciclista. Per moltes raons: per la seva mida no excessivament gran, per la seva condició de ciutat pràcticament plana, sense desnivells; per les condicions climàtiques agradables i pel carril bici de més de 120 quilòmetres que connecta tots els barris de la ciutat. A més, l´anell ciclista envolta el centre històric de València i permet accedir fàcilment a aquesta zona absolutament clau de la ciutat.\n\nValència pertany a la Xarxa de Ciutats per la Bicicleta, associació per a la generació d'una dinàmica entre ciutats espanyoles que permeti facilitar, fer més segura i desenvolupar la circulació dels ciclistes, especialment al medi urbà.")

indices = [18, 10, 20, 8, 7, 14, 13, 15, 16, 21, 0, 2, 1, 6, 22, 5, 19, 3, 12, 11, 23, 9, 4, 17]
```

```python
df_3 = pd.DataFrame({'descripcion':descripcion, 'indices':indices})

df_3 = df_3.sort_values(by='indices')

descripcion_2 = list(df_3.descripcion)
```

```python
df_rutas['description'] = descripcion_2

df_rutas.description
```

```python
df_rutas.to_csv('data/routes_clean.csv')
```
