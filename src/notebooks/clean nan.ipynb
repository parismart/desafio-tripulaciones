```python
import pandas as pd
import numpy as np
```

```python
routes = pd.read_csv("data/definitivo_routes.csv")
```

```python
routes
```

```python
routes["dificulty"] = routes["dificulty"].fillna("")
```

```python
routes["duration"] = routes["duration"].fillna(0)
```

```python
routes["duration"] = routes["duration"].astype("int64")
```

```python
routes.to_csv("data/absolute_routes.csv", index= False)
```
