```python
full_df["list_coor"] = full_df["coordenadas"].str.replace("[","").str.replace("]","").str.split(",")
```

```python
full_df["long"] = full_df["list_coor"].str[1]
```

```python
full_df["long"] = full_df["long"].astype("float32")
```

```python
full_df = full_df.drop(columns=["coordenadas","list_coor"])
```

```python
full_df = full_df.drop(columns="descripcion_ruta")
```

```python
full_df
```

```python
routes = list(full_df["nombre_ruta"].unique())
```

```python
routes.index("Ruta València amb bicicleta")
```

```python
def generate_route_id(route):
    id = routes.index(route)
    return id
```

```python
full_df["id_route"] = full_df["nombre_ruta"].apply(generate_route_id)
```

```python
colums =['id_route','nombre_ruta', 'nombre_punto_interes', 'descripcion', 'lat', 'long']
```

# Combinar Traducciones

```python
rutas_df = pd.read_csv("data/supreme_routes.csv")
```

```python
# rutas_df
```

```python
routes_trans = pd.read_csv("data/routes_translated.csv").drop("Unnamed: 0", axis=1)
```

```python
routes_trans = routes_trans.iloc[:,-4:]
```

```python
# routes_trans
```

```python
full_rutas = pd.concat([rutas_df, routes_trans], axis= 1)
```

```python
full_rutas
```

```python
full_rutas = full_rutas.rename(columns={"description": "val_description", "resume": "esp_resume", "castDescrip": "esp_description", "engDescrip": "eng_description" })
```

```python
# full_rutas
```

```python
list(full_rutas.columns)
```

```python
cols = ['route_id',
 'name',
 'esp_resume',
 'eng_resume',
 'val_resume',
 'duration',
 'dificulty',
 'start',
 'end',
 'image',
 'url',
 'transport',
 'type',
 'esp_description',
 'val_description',
 'eng_description']
```

```python
full_rutas = full_rutas[cols]
```

```python
full_rutas.to_csv("data/ultimate_routes.csv", index = False)
```
