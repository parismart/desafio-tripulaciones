```python
import pandas as pd
```

```python
df = pd.read_csv("data/puntos_interes.csv")
```

```python
df = df.drop("Unnamed: 0", axis = 1)
```

```python
extra_df = pd.read_csv("data/extra_df.csv")
```

```python
# extra_df
```

```python
extra_df = extra_df.rename(columns={"descripcion_ruta": "descripcion", "descripcion": "descripcion_ruta"})
```

```python
cols = list(extra_df.columns)
cols
```

```python
cols = ['descripcion',
 'coordenadas',
 'nombre_ruta',
 'nombre_punto_interes',
 'descripcion_ruta']
```

```python
extra_df = extra_df[cols]
```

```python
full_df = pd.concat([df,extra_df]).reset_index(drop=True)
```

```python
full_df
```

```python
cols = ['nombre_ruta', 'nombre_punto_interes', 'descripcion', 'coordenadas','descripcion_ruta']
```

```python
full_df = full_df[cols]
```

```python
full_df
```

```python
full_df["list_coor"] = full_df["coordenadas"].str.replace("[","").str.replace("]","").str.split(",")
```

```python
full_df["lat"] = full_df["list_coor"].str[0].str.replace("'","")
```

```python
full_df["lat"] = full_df["lat"].astype("float32")
```

```python
full_df["long"] = full_df["list_coor"].str[1]
```

```python
full_df["long"] = full_df["long"].astype("float32")
```

```python
full_df = full_df.drop(columns=["coordenadas","list_coor"])
```

```python
full_df = full_df.drop(columns="descripcion_ruta")
```
